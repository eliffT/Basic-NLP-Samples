# Stop Words


---



***Stop Words, temel olarak gereksiz kelimelerdir. Cümle akışını güzelleştirmek için kullanılan ancak veri analizinde anlam ifade etmeyen kelimelerdir.***

***Doğal dil işlemede kullanılan metine corpus denilir.***

```python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text = "Geçmişte doğal dil için tamamen kural tabanlı sistemler kullanılmaktaydı. 1960'larda MIT' de tamamen kural tabanlı çalışan Eliza isminde bir chatbot geliştirilmiştir."
```

```python
import nltk
nltk.download('stopwords')
```

***NLtk' de bulunan ingilizce stopwordsler için kullanılır.***

```python

stopwords = stopwords.words('turkish')
```

```python
words = word_tokenize(text)
```

```python
filtered_words = []
for word in words:
    if word not in stopwords:
        filtered_words.append(word)
        
```
