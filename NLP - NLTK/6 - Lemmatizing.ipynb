**Kök bulmak için kullanılır.
Stemming işleminde kelimenin sadece sondaki ekleri kesiliyor. Ekstra bir işlem yapılmaz.
Lemmatizing ile gerçekten kelimenin kökü bulunabilir. Bunun kelime hazneleri ve kelimelerin morfolojik analizleri kullanılır. Bu sayede kelime kökleri daha doğru bir şekilde bulunabilmektedir.**

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lem = WordNetLemmatizer()
```

```python
words = ['drive', 'driving', 'driver', 'drives', 'drove', 'cats','children']
```

```python
for w in words:
    print(lem.lemmatize(w))
```

```python
#'v' eklendiğinde kelimenin verb- fiil olduğu bilgisi verilir. Bu şekilde kök alma işlemi gerçekleşir.
lem.lemmatize('drove','v')
```
