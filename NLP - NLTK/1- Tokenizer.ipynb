---



```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
text = 'Alan Mathison Turing, İngiliz matematikçi, bilgisayar bilimcisi ve kriptolog. Bilgisayar biliminin kurucusu sayılır. \
Geliştirmiş olduğu Turing testi ile makinelerin ve bilgisayarların düşünme yetisine sahip olup olamayacakları konusunda bir kriter öne sürmüştür.'
```

***Python .split metodu kullanılarak tokenlerine ayrılabilir.***
***Fakat noktalama işaretleri bazı kelimelerde dahil edilirken bazılarında dahil edilmemiştir.***

```python

text.split()
```


***Nltk kütüphanesinin word_tokenize fonksiyonu ile tokenlarına ayrılmış oldu. NOk. ayrı token olmuştur.*** 

```python

word_tokenize(text)
```

***Cümleler token olarak ayrılmıştır. liste olarak değer elde edilir. Liste 
özellikleri kullanılır.***

```python

sent_tokenize(text)
```

```python
for token in word_tokenize(text):
    print(token)
```
